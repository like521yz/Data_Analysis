```python
import pandas as pd
```

```python
# Series 的创建
s = pd.Series([1,2,3,4])
print(s)
```

```python
# 自定义索引
arr_s = pd.Series([1,2,3,4],index = ['A','B','C','D'])
print(arr_s)
# 定义name
arr_s2 = pd.Series([1,2,3,4],index = ['A','B','C','D'],name = '月份')
print(arr_s2)
```

```python
#通过索引裁剪
arr_s = pd.Series([1,2,3,4],index = ['A','B','C','D'])
s1 = pd.Series(arr_s,index = ['A','B'])
print(s1)
```

```python
print(arr_s.loc['A'])       # 按标签取单个值
print("--------------------")
print(arr_s.iloc[0])        # 按位置取单个值（批量用iloc[[0,1]]）
print("--------------------")
print(arr_s.at['A'])        # 按标签快速取单个值
print("--------------------")
print(arr_s.iat[0])      # 按位置快速取单个值（批量用iat不行）
```

```python
# 访问数据
arr = pd.Series([1,2,3,4],index = ['A','B','C','D'])
print(arr['A'])
print(arr[arr<3])
print(arr.head(2))
print(arr.tail(1))
```

#### 常用方法

```python
s = pd.Series([None,1,2,3,4,5],index=['A','B','C','D','E','F'])
print(s)
```

```python
print(s.mean())# 获取平均值
print("-------------------")
print(s.std)# 获取标准差
print("-------------------")
print(s.var())# 获取方差
print("-------------------")
print(s.sum())# 求和
print("-------------------")
print(s.max())# 最大值
print("-------------------")
print(s.min())# 最小值
print("-------------------")
print(s.median())# 中位数（去掉NaN值）
print("-------------------")
print(s.quantile(0.8))# 分位数
#-------------
#1  2  3  4  5
#位置：0.8*4 = 3.2
#值：4 + (5-4) * 0.2 = 4.2

```

```python

print(s.mode())# 众数
print("-------------------")
print(s.value_counts())# 每个元素的计数
print("-------------------")
```

```python
# 默认取前五行
print(s.head(2))
print("-------------------")
# 默认取后五行
print(s.tail(1))
print("-------------------")
# 
```

```python
# 查看所有描述性信息
print(s.describe())
```

```python
# 获取元素个数
print(s.count())
print("------------------")
# 获取索引
print(s.keys())
print(s.index)
print("------------------")

```

```python
# 判断元素是否缺失值 True 为 '缺失'
print(s.isna())
print("------------------")
# 检查元素是否在参数列表中 True 为 '存在'
print(s.isin([4,5,6]))
```

```python
# 去重
s = pd.Series([1,1,2,3,4],index = ['A','B','C','D','E'])
print(s.drop_duplicates())
print("--------------------")
print(s.unique())
```

```python
s = pd.Series([4,1,5,3,0],index = ['A','B','C','D','E'])
print(s.sort_values())# 数值从小到大排序
print("-------------------")
print(s.sort_index())# 按照索引排序
```

```python
# 分析股票数据
import pandas as pd

# 构造收盘价Series，索引为连续10天的日期
prices = pd.Series(
    [102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
    index=pd.date_range('2023-01-01', periods=10)
)

print("收盘价数据：")
print(prices)
```

```python
# 计算收益率
daily_returns = prices.pct_change()
print(daily_returns)
```

```python
# 计算收益率最高和最低日期
print(f"收益率最高日期为：{daily_returns.idxmax()},收益率为：{daily_returns[daily_returns.idxmax()]}")
print(f"收益率最低日期为：{daily_returns.idxmin()},收益率为：{daily_returns[daily_returns.idxmin()]}")
```

```python
# 计算波动率（标准差），自动忽略NaN值
volatility = daily_returns.std()
print(f"波动率（收益率标准差）：{volatility:.6f}")
```
