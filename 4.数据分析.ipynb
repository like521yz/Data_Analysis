```python
# 缺失值处理
import numpy as np
import pandas as pd
data = {
    "姓名": ["张三", "李四", None, "王五", "赵六", "钱七", "孙八"],
    "年龄": [18, None, 19, 20, 18, None, 19],
    "数学成绩": [90, 85, 78, None, 92, 88, 76],
    "语文成绩": [88, 92, None, 85, 90, 87, 89]
}
df = pd.DataFrame(data)
print("=== 原始数据集 ===")
df
print("========统计每列缺失值的数量=========")
print(df.isnull().sum())
print("=========快速查看数据概况==========")
df.info()
```

```python
# 剔除缺失值
## 方式1：剔除任意一列有缺失值的行（最常用） 
print("========剔除缺失值后的数据==========")
df_drop_any = df.dropna(how="any")# how="any"：只要行中有1个缺失值就删除
df_drop_any
```

```python
## 方式2：仅剔除“姓名”列有缺失值的行（指定列剔除，更灵活）
df_drop_subset = df.dropna(subset=["数学成绩"])
print("\n=== 仅剔除姓名列有缺失值后的数据集 ===")
df_drop_subset
```

```python
print("\n=== 验证：剔除后数据的缺失值数量 ===")
print(df_drop_any.isnull().sum())
```

```python
# 处理重复值
import pandas as pd

# 1. 构造带重复值的示例数据（模拟电商订单数据）
data = {
    "订单号": ["001", "002", "001", "003", "002", "004", "005", "005"],
    "用户ID": ["U001", "U002", "U001", "U003", "U002", "U004", "U005", "U005"],
    "商品": ["手机", "电脑", "手机", "平板", "电脑", "耳机", "手表", "手表"],
    "金额": [5999, 7999, 5999, 2999, 7999, 399, 1999, 1999],
    "支付时间": ["2026-02-01 10:00", "2026-02-01 11:00", "2026-02-01 10:00",
                "2026-02-01 12:00", "2026-02-01 11:00", "2026-02-01 13:00",
                "2026-02-01 14:00", "2026-02-01 14:01"]  # 005订单有2条，仅时间差1分钟
}
df = pd.DataFrame(data)
# 识别重复值
df["是否重复值"] = df.duplicated()
#df
# 统计重复值数量（按所有列）
duplicate_count = df.duplicated().sum()
print(f"\n=== 所有列完全重复的行数：{duplicate_count} ===")
```

```python
# 按指定列判断重复
df["订单号是否重复"] = df.duplicated(subset=["订单号"])
df[["订单号","用户ID","商品","订单号是否重复"]]
```

```python
# 删除所有列完全重复的行(只保留首次出现)
df_drop_all = df.drop_duplicates()
df_drop_all
```

```python
# 按指定列删除重复
df_drop_order = df.drop_duplicates(subset=["订单号"])
df_drop_order[["订单号","用户ID","支付时间","订单号是否重复"]]
```

```python
# 按指定列去重，保留最后一次出现的行（比如保留最新支付时间)
df_drop_last = df.drop_duplicates(subset=["订单号"],keep="last")
df_drop_last[["订单号","用户ID","支付时间","订单号是否重复"]]
```

```python
# 彻底删除所有重复行（包括首次出现的，仅保留唯一行）
df_drop_all_duplicate = df.drop_duplicates(subset=["订单号"],keep=False)
df_drop_all_duplicate
```

```python
# 数据变形
import pandas as pd

# 1. 构造初始宽表数据（模拟月度销售数据）
# 宽表特点：列是类别（月份），行是维度（产品+地区）
data = {
    "产品": ["手机", "手机", "电脑", "电脑"],
    "地区": ["华东", "华北", "华东", "华北"],
    "1月销量": [500, 300, 200, 150],
    "2月销量": [550, 320, 210, 160],
    "3月销量": [600, 350, 220, 170]
}
df_wide = pd.DataFrame(data)
df_wide
```

```python
# 宽表转长表（melt）：把多列（月份）转为“键-值”对
df_long = df_wide.melt(
    id_vars=["产品", "地区"],  # 保留不变的标识列
    value_vars=["1月销量", "2月销量", "3月销量"],  # 要“拉长”的列
    var_name="月份",  # 原列名转为新列的名称
    value_name="销量"  # 原列值转为新列的名称
)
df_long
```

```python
# 长表转宽表（pivot）：把“键-值”对转回多列
df_pivot = df_long.pivot(
    index=["产品","地区"],
    columns=["月份"],# 列维度（把“月份”的不同值转为列）
    values=["销量"]# 填充到单元格的值
).reset_index()
df_pivot.columns.name = None
df_pivot
```

```python
# 行列转置（T）：行和列互换（适合快速调整数据展示形式）
df_transpose = df_wide.set_index(["产品","地区"]).T
df_transpose
```

```python
#进阶：pivot_table（处理重复值的长表转宽）
# 构造带重复值的长表（模拟同一产品+地区+月份有2条销量记录）
df_long_dup = pd.concat([df_long, pd.DataFrame({
    "产品": ["手机", "电脑"],
    "地区": ["华东", "华北"],
    "月份": ["1月销量", "2月销量"],
    "销量": [50, 20]
})], ignore_index=True)
#df_long_dup
## 用pivot_table聚合重复值（比如求和）
df_pivot_table = df_long_dup.pivot_table(
    index=["产品","地区"],
    columns=["月份"],
    values=["销量"],
    aggfunc=sum
) .reset_index()
df_pivot_table
```
